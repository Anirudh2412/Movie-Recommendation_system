Feature Selection Part

```python
movies.columns
```

```python
movies=movies[['id', 'title', 'overview', 'genres']]
```

```python
movies
```

```python
movies['tags'] = movies['overview']+movies['genres']
```

```python
movies
```

```python
new_data  = movies.drop(columns=['overview', 'genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv=TfidfVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
import difflib
```

```python
def recommand():
    movie_name = input(' Enter your favourite movie name : ')
    list_of_all_titles = new_data['title'].tolist()
    find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
    close_match = find_close_match[0]
    index=new_data[new_data['title']==close_match].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand()
```
